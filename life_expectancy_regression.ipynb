### 기대 수명 회귀 예측

#### Features

- Country: 국가
- Year: 연도
- Status: 개발도상국(Developing) / 선진국(Developed)
- Adult Mortality: 성인 사망률 (1000명 기준 15-60세 인구의 사망 확률)
- infant deaths: 유아 사망자 수 (1000명 당)
- Alcohol: 1인 당 알코올 섭취량 (순수 알코올, 리터 기준)
- percentage expenditure: GDP 중 건강 관련 지출액의 비중
- Hepatitis B: 1세 미만 영유아 중 헵티티스 B 백신 접종률
- Measles: 1000명 당 보고된 홍역 발병 건수
- BMI: 평균 BMI 지수
- under-five deaths: 1000명 당 5세 이하 사망자 수
- Polio: 1세 미만 영유아 중 폴리오(Pol3) 백신 접종률
- Total expenditure: 총 정부 지출 중 건강 지출 비중
- Diphtheria: 1세 미만의 영유아 중 디프테리아, 파상풍, 백일해(DTP) 예방접종률
- HIV/AIDS: 0~4세의 아이들 중 HIV/AIDS로 인한 1000명 당 사망자 수
- GDP: GDP
- Population: 국가 총 인구
- thinness 1-19 years: 10세에서 19세 사이의 아이들과 청소년들 사이에서 얇음 증후군의 유병률
- thinness 5-9 years: 5세에서 9세 사이의 아이들 사이에서 얇음 증후군의 유병률
- Income composition of resources: 소득 자원의 구성에 따른 인간 개발 지수(HDI)
- Schooling: 평균적인 교육 년수
